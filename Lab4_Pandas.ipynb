```python
import pandas as pd
import numpy as np


# КРОКИ 1-3: Створення, перевірка типу, values та index
names_list = ["Bob", "Emma", "Joe", "Lili", "Anna", "Alex", "Max", "Mia", "Leo", "Eva", "Tim", "Zoe", "Dan", "Joy", "Ben"]
ds1 = pd.Series(names_list)

print("1. Об'єкт ds1:")
print(ds1)

print(f"\n2. Тип об'єкта ds1: {type(ds1)}")  # Перевірка типу

print("\n3. Значення Series (ds1.values):")
print(ds1.values)  # Повертає масив NumPy

print("\nНабір індексів (ds1.index):")
print(ds1.index)  # Повертає RangeIndex


# КРОКИ 4-5: Вибірка та зрізи (Неявні / Позиційні індекси)

print("\n4. Перші два рядки ds1 (через head або неявний зріз):")
print(ds1.iloc[0:2])  # Або ds1.head(2). Кінцевий індекс виключається

print("\n5. Непарні рядки у діапазоні з 3 по 9 рядок:")
print(ds1.iloc[3:10:2])  # Виведе елементи на позиціях 3, 5, 7, 9
# КРОКИ 6-7: Робота з явними індексами (Стиль словника)
# Попереднє створення словника відповідно до рисунка в умові
custom_info_dict = {
    "Jan": 125,
    "Feb": 234,
    "Mar": 13,
    "Apr": 356,
    "May": 87
}
custom_info = pd.Series(custom_info_dict)
print("\n6. Об'єкт custom_info (створений зі словника):")
print(custom_info)

print("\n7. Тривалість розмов у весняні місяці (одним оператором):")
# Використовуємо розширене (fancy) індексування за явними мітками
print(custom_info[['Mar', 'Apr', 'May']])

# КРОКИ 8-9: Переіндексація та виявлення пропущених даних (NaN)
m_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul']

# Створюємо новий Series. За правилами вирівнювання з методички,
# нові індекси 'Jun' та 'Jul', яких не було в custom_info, автоматично заповняться NaN.
custom_info_new = pd.Series(custom_info, index=m_names)
print("\n8. Об'єкт custom_info_new:")
print(custom_info_new)

print("\n9. Виявлення пропущених даних (missing values):")
print(custom_info_new.isna())  # Альтернатива: custom_info_new.isnull()


# КРОКИ 10-12: Зрізи за мітками, фільтрація та модифікація даних
print("\n10. Зріз даних від березня (Mar) до травня (May):")
# кінцевий індекс ВКЛЮЧАЄТЬСЯ в зріз.
print(custom_info_new['Mar':'May'])

print("\n11. Рядки, де тривалість дзвінка < 100:")
# Застосування маски фільтрації за умовою
print(custom_info_new[custom_info_new < 100])

print("\n12. Помножити дані із тривалістю дзвінка < 200 на 5:")
# Створюємо копію, щоб не зіпсувати оригінальний custom_info_new
custom_info_mult = custom_info_new.copy()
# Модифікуємо значення за маскою (елементи з NaN і ті, що >= 200, залишаться незмінними)
custom_info_mult[custom_info_mult < 200] *= 5
print(custom_info_mult)
```
